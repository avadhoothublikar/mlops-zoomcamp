scenario 2: A cross-functional team with one data scientist working on an ML model

MLflow setup:

tracking server: yes (local)
backend store: sqlite db
artifacts store: local filesystem

The experiments can be accessed locally by accessing the local tracking server
To run this example you need to launch mlflow server locally by running the following command in terminal

mlflow server --backend-store-uri sqlite:///mlflow.db --default-artifact-root ./artifacts_local

```python
import mlflow

mlflow.set_tracking_uri("http://127.0.0.1:5000")
```

```python
print(f"tracking uri: {mlflow.get_tracking_uri()}")
```

```python
mlflow.search_experiments()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris

mlflow.set_experiment('my-experiment-1')

with mlflow.start_run():

    X, y = load_iris(return_X_y=True)

    params = {'C': 1.0, 'random_state': 42}
    mlflow.log_params(params)

    lr = LogisticRegression(**params).fit(X, y)
    y_pred = lr.predict(X)
    mlflow.log_metric('accuracy', accuracy_score(y, y_pred))

    mlflow.sklearn.log_model(lr, artifact_path='models')
    print(f"default artifact_uri: '{mlflow.get_artifact_uri()}'")
```

```python
mlflow.search_experiments()
```

```python
from mlflow.tracking import MlflowClient

client = MlflowClient("http://127.0.0.1:5000")
```

```python
client.search_registered_models()
```

```python
run_id = client.search_runs(experiment_ids=['1'])[0].info.run_id
mlflow.register_model(
    model_uri=f"runs:/{run_id}/models", 
    name="iris-classifier"
)
```
