```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
```

```python
#Importing March 2023 Yellow taxi trips data

url = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-03.parquet"
df = pd.read_parquet(url)

#Number of records in the dataset
num_records = len(df)
print(f"Number of records in the dataset: {num_records}")

# Applied same preprocessing logic
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
df['duration'] = df['duration'].dt.total_seconds() / 60

df = df[(df['duration'] >= 1) & (df['duration'] <= 60)]

categorical = ['PULocationID', 'DOLocationID']
df[categorical] = df[categorical].astype(str)

# Resulting dataset size
num_records_after_preprocessing = len(df)
print(f"Number of records after preprocessing: {num_records_after_preprocessing}")
```

```python
#prepares the features and target
features = ['PULocationID', 'DOLocationID']
target = 'duration'

train_dicts = df[features].to_dict(orient='records')
y_train = df[target].values

#DictVectorizer
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)

#Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)

#print inercept
lr.intercept_
```

```python
import mlflow
import mlflow.sklearn

mlflow.set_tracking_uri("http://localhost:5000")
mlflow.set_experiment("homework-04")

with mlflow.start_run():
    mlflow.sklearn.log_model(lr, artifact_path="model")
```
